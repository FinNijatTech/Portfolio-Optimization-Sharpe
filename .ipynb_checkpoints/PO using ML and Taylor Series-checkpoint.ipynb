```python
import matplotlib.pyplot as plt

# Plot the actual and predicted returns
plt.scatter(y_train, returns_model.predict(X_train), label='train')
plt.scatter(y_test, returns_model.predict(X_test), label='test')
plt.xlabel('Actual Returns')
plt.ylabel('Predicted Returns')
plt.legend()
plt.show()

# Plot the actual and predicted risks
plt.scatter(y_train, risk_model.predict(X_train), label='train')
plt.scatter(y_test, risk_model.predict(X_test), label='test')
plt.xlabel('Actual Risks')
plt.ylabel('Predicted Risks')
plt.legend()
plt.show()
```
